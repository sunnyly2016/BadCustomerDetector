```python
from BadCustomerDetector.BadCustomerDetection import BadCustomerDetection as bcd
import pandas as pd
```

```python
df = pd.read_csv('./test/unique_consumers.csv')
```

```python
detector = bcd()
```

```python
outliers = detector.bad_customer_detector(df, method = 'KNN')
```

```python
outliers.groupby(['Cluster','bad_customer']).agg(
    # Get max of the duration column for each group
    Avg_unique_clicked=('unique_offer_clicked', 'mean'),
    Avg_total_clicks=('total_offer_clicks', 'mean'),
    Avg_unique_rides=('unique_offer_rides', 'mean'),
    Avg_total_rides=('total_offer_rides', 'mean'),
    Avg_total_claimed=('total_offers_claimed', 'mean'),
    bad_customer_counts = ('bad_customer', 'count')
    # Get min of the duration column for each group
).round(1)
```

```python
outliers = detector.bad_customer_detector(df, method = 'Voting')
```
